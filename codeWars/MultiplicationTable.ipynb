```polyglot-notebook
let multiplicationTable (n: int): int list list =
    let header = [1..n]
    List.map (fun i -> List.map (fun j -> i * j) header) header
```

```polyglot-notebook

let multiplicationTable2 (n: int): int list list  =
    [ for i in 1..n do yield [i..i..(i*n)] ]
```

```polyglot-notebook
multiplicationTable2 4
```
